## Utilice el m√©todo de Euler para aproximar las soluciones para cada uno de los siguientes problemas de valor inicial.
![](tarea12.1.png)

### a)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
def euler_method(f, t0, y0, h, t_end):
    t_values = [t0]
    y_values = [y0]
    t = t0
    y = y0
    while t < t_end:
        y = y + h * f(t, y)
        t = round(t + h, 10)  
        t_values.append(t)
        y_values.append(y)
    return t_values, y_values
def f1(t, y):
    return y / t - (y / t) ** 2
t0, y0, h, t_end = 1, 1, 0.1, 2
t_vals, y_vals = euler_method(f1, t0, y0, h, t_end)
df = pd.DataFrame({"t": t_vals, "y": y_vals})
print(df)
plt.plot(t_vals, y_vals, '-s', label='aproximacion')
plt.show()
```

### b)

```python
def f1(t, y):
    return 1 + y/t + (y/t)**2 
t0, y0, h, t_end = 1,0,0.2,3
t_vals, y_vals = euler_method(f1, t0, y0, h, t_end)
df = pd.DataFrame({"t": t_vals, "y": y_vals})
print(df)
plt.plot(t_vals, y_vals, '-s', label='aproximacion')
plt.show()
```

### c)

```python
def f1(t, y):
    return -(y+1)*(y+3)
t0, y0, h, t_end = 0,-2,0.2,2
t_vals, y_vals = euler_method(f1, t0, y0, h, t_end)
df = pd.DataFrame({"t": t_vals, "y": y_vals})
print(df)
plt.plot(t_vals, y_vals, '-s', label='aproximacion')
plt.show()
```

### d)

```python
def f1(t, y):
    return -5*y + 5*t**2+2*t
t0, y0, h, t_end = 0,1/3,0.1,1
t_vals, y_vals = euler_method(f1, t0, y0, h, t_end)
df = pd.DataFrame({"t": t_vals, "y": y_vals})
print(df)
plt.plot(t_vals, y_vals, '-s', label='aproximacion')
plt.show()
```

## Aqu√≠ se dan las soluciones reales para los problemas de valor inicial en el ejercicio 3. calcule el error real en las aproximaciones del ejercicio 3.
![](tarea12.2.png)

### a)

```python
def f1(t, y):
    return y / t - (y / t) ** 2
def exacto(t):
    return t / (1 + np.log(t))
t0, y0, h, t_end = 1, 1, 0.1, 2
t_vals, y_vals = euler_method(f1, t0, y0, h, t_end)
df = pd.DataFrame({"t": t_vals, "y": y_vals})
df["Exacto y"] = df["t"].apply(exacto)
df["Error"] = abs(df["y"] - df["Exacto y"])
print(df)
plt.plot(t_vals, y_vals,'-s',label='Aproximacion',color='Blue')
plt.plot(t_vals, exacto(t_vals),'-s',label='Real',color='Red')
plt.show()
```
